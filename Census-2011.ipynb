## Finding latitude and longitude of every district  by using geopy library

```python
districts_coord = []
failed_location=[]
for i,j in zip(df['District name'],df['State name']):
    try:
        location=geolocator.geocode(i+ "," +j+ ", India")
        districts_coord.append([location.latitude, location.longitude])
    except AttributeError:
        failed_location.append(i)
        print(f"Failed to find location for {i}")
```

```python
def lat_long(data):
    latitude_dist=[]
    longitude_dist=[]
    for i,j in zip(data['District name'],data['State name']):
        try:
            val=geolocator.geocode(i+ "," +j+ ", India")
            latitude_dist.append(val.latitude)
            longitude_dist.append(val.longitude)
        except AttributeError:
            latitude_dist.append(np.nan)
            longitude_dist.append(np.nan)
            
        
    data['latitude']=latitude_dist
    data['longitude']=longitude_dist
    return data
```

### State wise Analysis

```python
# Calculating the percentage of every parameter with respect to population for every states in order to do analysis based on states

states_df=data.drop(columns=['District code','District name','latitude','longitude']).groupby('State name').sum().reset_index()
states_df
```

```python
## importing states geojson file
## to read the geojson file importing geopandas

import geopandas as gpd

states_geo = gpd.read_file('india_states.geojson')
states_geo
```

```python
# we will have to match the names of states according to the names present in states_geo in order to plot the co-ordinaates of each states

states_df['State name']=states_df['State name'].apply(lambda x:x.title())
```

### correcting the names of the state

```python
### applying numpy functions to replace the names

np.setdiff1d(states_df['State name'].unique(),states_geo['NAME_1'].unique())
```

```python
np.setdiff1d(states_geo['NAME_1'].unique(),states_df['State name'].unique())
```

```python
#replace

states_df['State name']=states_df['State name'].replace(['Andaman And Nicobar Islands', 'Dadra And Nagar Haveli','Daman And Diu', 'Jammu And Kashmir', 'Nct Of Delhi', 'Orissa','Pondicherry'],
                            ['Andaman and Nicobar Islands', 'Dadra and Nagar Haveli','Daman and Diu','Jammu and Kashmir','Delhi', 'Odisha','Puducherry'])
```

```python
states_df
```

```python
## converting all the parameter values into percentage with respect to population 
def percent_state(data):
    cols=data.columns[2:]
    for i in cols:
        data[i]=np.round(data[i]/data['Population']*100,2)
        
    return data
```

```python
final_df=percent_state(states_df)
final_df
```
